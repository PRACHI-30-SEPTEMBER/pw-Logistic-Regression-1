### Q1. Explain the difference between linear regression and logistic regression models. Provide an example of a scenario where logistic regression would be more appropriate.

Linear regression predicts continuous values, while logistic regression predicts probabilities for classification problems.
Example: Email spam detection (spam or not spam) → logistic regression is more appropriate.

### Q2. What is the cost function used in logistic regression, and how is it optimized?

Logistic regression uses the Log Loss (Binary Cross-Entropy) cost function.
It is optimized using techniques like Gradient Descent to minimize the error.

### Q3. Explain the concept of regularization in logistic regression and how it helps prevent overfitting.

Regularization adds a penalty to the loss function:
- L1 (Lasso): makes some coefficients zero
- L2 (Ridge): shrinks coefficients

It prevents overfitting by reducing model complexity.

### Q4. What is the ROC curve, and how is it used to evaluate the performance of the logistic regression model?

ROC (Receiver Operating Characteristic) curve plots True Positive Rate vs False Positive Rate.
AUC (Area Under Curve) measures performance:
- AUC = 1 → perfect model
- AUC = 0.5 → random model

### Q5. What are some common techniques for feature selection in logistic regression? How do these techniques help improve the model's performance?

- L1 Regularization (Lasso)
- Recursive Feature Elimination (RFE)
- Correlation-based selection

These remove irrelevant features, improving accuracy and reducing overfitting.

### Q6. How can you handle imbalanced datasets in logistic regression? What are some strategies for dealing with class imbalance?

- Oversampling (SMOTE)
- Undersampling
- Class weight adjustment
- Using evaluation metrics like F1-score instead of accuracy

### Q7. Can you discuss some common issues and challenges that may arise when implementing logistic regression, and how they can be addressed? For example, what can be done if there is multicollinearity among the independent variables?

Common issues:
- Multicollinearity → Use VIF, remove correlated features, or use regularization
- Overfitting → Use regularization
- Imbalanced data → Resampling techniques
- Non-linearity → Feature transformation
